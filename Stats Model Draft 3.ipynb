```python
#Dropped rows where weeks out <5
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
from sklearn.preprocessing import LabelEncoder
lb_make = LabelEncoder()
```

```python
df2 = pd.read_csv('Gaming_Data_Cleaned_Draft1')
df2 = df2.drop(columns='Unnamed: 0')
```

```python
weekly_sales = list(df2.groupby('Game Title').mean()['Weekly Sales'])
```

```python
all_time_sales = list(df2.groupby('Game Title').max()['All Time Sales'])
```

```python
weeks_out = list(df2.groupby('Game Title').max()['Weeks Out'])
```

```python
genre = list(df2.groupby('Game Title').max()['Genre'])
```

```python
console = list(df2.groupby('Game Title').max()['Console'])
```

```python
publisher = list(df2.groupby('Game Title').max()['Publisher'])
```

```python
df2 = pd.DataFrame(columns=['Weekly_Sales', 'All_Time_Sales', 'Weeks_Out', 'Genre', 'Console', 'Publisher'])
```

```python
df2.Weekly_Sales = weekly_sales
df2.All_Time_Sales = all_time_sales
df2.Weeks_Out = weeks_out
df2.Genre = genre
df2.Console = console
df2.Publisher = publisher
```

```python
df2.info()
```

```python
df2 = df2[df2['Weeks_Out'] >= 5]
```

```python
pub_dummies = pd.get_dummies(df2["Publisher"], prefix="pub")
gen_dummies = pd.get_dummies(df2["Genre"], prefix="gen")
con_dummies = pd.get_dummies(df2["Console"], prefix="con")
```

```python
df2 = df2.drop(["Publisher", "Genre", "Console"], axis=1)
df2 = pd.concat([df2, pub_dummies, gen_dummies, con_dummies], axis=1)
df2.head()
```
