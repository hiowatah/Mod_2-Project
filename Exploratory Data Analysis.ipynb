```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify
```

```python
df = pd.read_csv('Gaming_Data_Cleaned_Draft1.csv')
df = df.drop(columns='Unnamed: 0')
```

```python
df.head()
```

```python
fig = plt.gcf()
ax = fig.add_subplot()
fig.set_size_inches(15, 10)

squarify.plot(label=list(df.groupby('Console').count().index),
              sizes=list(df.groupby('Console').count().Genre),color=["red","orange","yellow","green",
                                                                     "blue","purple","brown","lightblue",
                                                                     "pink"],alpha=.6);
plt.title('Software Sales by Hardware 2017-2018');
```

```python
df.groupby('Publisher').count().sort_values(by=['Game Title'],
                                            ascending = False).head(10)['Game Title'].plot(kind = 'barh');
plt.title('Most Active Publishers Between 2017 and 2018');
```

```python
df.groupby('Genre').count().sort_values(by=['Console'],
                                            ascending = False).head(10)['Console'].plot(kind = 'barh');
plt.title('Most Popular Genre Between 2017 and 2018');
```

```python
plt.scatter(df['Weeks Out'],df['All Time Sales'], label = 'Weeks Out');
plt.legend();
plt.xlabel('Number of Weeks Since Release')
plt.ylabel('Sales in Ten Millions')
plt.xlim(0, 200)
plt.title('Linearity Check')
```
